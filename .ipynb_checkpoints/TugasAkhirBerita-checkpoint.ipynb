```python
import copy
import numpy as np
import pandas as pd
from numpy import inf
import importlib as imp
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
factory_stemmer = StemmerFactory()
factory_stopword = StopWordRemoverFactory()

np.set_printoptions(threshold=np.nan) #menampilkan semua data dengan numpy

stemmer = factory_stemmer.create_stemmer()
stopword = factory_stopword.create_stop_word_remover()
```

```python
dataset = pd.read_excel("Dataset_Klasifikasi_Multi-Label_Topik_Berita.xlsx").fillna(0).as_matrix() # mengambil data dan mengubah menjadi matriks
dataset = dataset[:27,:] # mengambil data set sampai kolom 27


for i in range(dataset[:,1].size):
    dataset[i,1] = dataset[i,1].replace('\n','') #menghilangkan enter
    dataset[i,1] = stopword.remove(dataset[i,1])
    dataset[i,1] = stemmer.stem(dataset[i,1])
#update kata dasar C:\Users\ottosiahaan\AppData\Local\Programs\Python\Python36-32\Lib\site-packages\Sastrawi\Stemmer\data
```

```python
feature = np.array([])# membuat array kosong
for i in range(dataset[:,1].size):
    feature = np.append(feature,np.unique(np.array(dataset[i,1].split(' '))))

#feature.size untuk melihat banyak data sebelum


```

```python
feature = np.unique(feature)
#feature.size menghitung data setelah di unik
#feature
```

```python
term_dataset = np.zeros((dataset[:,1].size, feature.size))
```

```python
for i in range(dataset[:,1].size):
    tmp = np.array(dataset[i,1].split(' '))
    for j in range(feature.size):
        term_dataset[i,j] = np.sum(tmp == feature[j])
#feature[]
```

```python
def tfWeighted(x): #variabel umum
    rslt = copy.deepcopy(x) 
    rslt = 1 + np.log10(x)
    rslt[rslt == -inf] = 0 # megubah inf jadi 0
    return rslt
#x = tfWeighted(term_dataset)
#x
```

```python
def idfWeighted(x):
    rslt = np.zeros(x.shape[1]) #ukuran array
    for i in range(rslt.size):
            rslt[i] = x.shape[0]/np.sum(x[:,i])
    return rslt
#y = idfWeighted(term_dataset)
#y
```

```python
def tfidfWeighted(tfW,idf):
    idf_table = copy.deepcopy(tfW)
    idf_table[:,:] = idf
    return tfW * idf_table
#z = tfidfWeighted(x,y)
#z
```

```python
def binlabel(y):
    rslt = np.zeros(shape = (y.shape[0],13))
    for i in range(y.shape[0]):
        for j in range(y.sahpe[1]):
            if y[i,j] != 0:
                rslt[i,y[i,j]] =1
    return rslt

#y = dataset[:,2:5]
#y
```
